```python
df2.rename(columns = {"2021":"사례수","2021.1":"음식/미식탐방", "2021.2":"쇼핑", "2021.3":"자연 풍경 감상", "2021.4":"패션 유행 등 세련된 현대 문화 체험",
                       "2021.5":"역사/문화유적/전통문화 체험", "2021.6":"유흥/놀이 시설 체험", "2021.7":"치료 및 미용 서비스 체험", 
                       "2021.8":"여행 가능한 여가시간", "2021.9":"K-POP/한류스타 관련 공연팬미팅촬영지 등 체험", 
                       "2021.10":"레저 및 스포츠 활동", "2021.11":"자국과의 이동거리", "2021.12":"경제적인 여행비용","2021.13":"기타"}, inplace=True)
df2
```

```python
data_year = df2.iloc[14:18:,1:]
data_year
```

```python
data_year.rename(columns = {"통계분류(2)":"분기별"}, inplace=True)
data_year
```

```python
data_year.columns[1:]
```

```python
data_year.set_index('분기별',inplace=True)
data_year = data_year.astype('float')   
data_year.reset_index(inplace=True)
data_year
```

```python
data_year_copy = data_year.copy()
data_year_copy['음식/미식탐방'] = data_year_copy['사례수']*data_year_copy['음식/미식탐방']/100
data_year_copy['쇼핑'] = data_year_copy['사례수']*data_year_copy['쇼핑']/100
data_year_copy['자연 풍경 감상'] = data_year_copy['사례수']*data_year_copy['자연 풍경 감상']/100
data_year_copy['패션 유행 등 세련된 현대 문화 체험'] = data_year_copy['사례수']*data_year_copy['패션 유행 등 세련된 현대 문화 체험']/100
data_year_copy['역사/문화유적/전통문화 체험'] = data_year_copy['사례수']*data_year_copy['역사/문화유적/전통문화 체험']/100
data_year_copy['유흥/놀이 시설 체험'] = data_year_copy['사례수']*data_year_copy['유흥/놀이 시설 체험']/100
data_year_copy['치료 및 미용 서비스 체험'] = data_year_copy['사례수']*data_year_copy['치료 및 미용 서비스 체험']/100
data_year_copy['여행 가능한 여가시간'] = data_year_copy['사례수']*data_year_copy['여행 가능한 여가시간']/100
data_year_copy['K-POP/한류스타 관련 공연팬미팅촬영지 등 체험'] = data_year_copy['사례수']*data_year_copy['K-POP/한류스타 관련 공연팬미팅촬영지 등 체험']/100
data_year_copy['레저 및 스포츠 활동'] = data_year_copy['사례수']*data_year_copy['레저 및 스포츠 활동']/100
data_year_copy['자국과의 이동거리'] = data_year_copy['사례수']*data_year_copy['자국과의 이동거리']/100
data_year_copy['경제적인 여행비용'] = data_year_copy['사례수']*data_year_copy['경제적인 여행비용']/100
data_year_copy['기타'] = data_year_copy['사례수']*data_year_copy['기타']/100
data_year_copy
```
